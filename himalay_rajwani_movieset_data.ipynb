# **Import files**

```python
import warnings
import pandas as pd
import matplotlib.pyplot as plt
```

# **Read files**

```python
movie1=pd.read_csv("IMDb movies.csv")
movie1
```

# Cleaning of the DataFrame Movie 1

```python
movie1.dropna(axis=1,how="all",inplace=True)
movie1.dropna(axis=0,how="all",inplace=True)
```

```python
data1=movie1["year"].value_counts()[:5]
data1.values
```

# Q1) Top 5 year in which maximum no of movies released?

```python
plt.figure(figsize=(11,5))
plt.bar(data1.index,data1.values)

plt.xlabel("YEAR",color="red",fontsize=20)
plt.ylabel("Maximum no of movies released",color="red",fontsize=20)

plt.show()
```

# Q2) What type of movie liked most in this data set in the world ? 

```python
#splitting the columns and get the result in the dic format in this function
li={}
def fun(ge):
    
    f=ge.split(',')
    
    for i in range(len(f)):
        
        f[i]=f[i].replace(" ","")
        if f[i] not in li:
            
            li[f[i]]=1
        else:
            
            li[f[i]]=li[f[i]]+1
                   
                   
    return li
    
 
    
    
movie1["genre"].apply(fun)
```

# Plotting the graph 

```python
#splitting the genre column and get the result in the dic format in this function
l1=li.keys()
l2=li.values()

#dataframe 
df=pd.DataFrame(l2,l1,columns=["Values"])
df.sort_values("Values",ascending=False)

plt.figure(figsize=(21,10))
plt.bar(df.index,df["Values"])

plt.xticks(color="Red",rotation=90)
plt.yticks(color="Red")

plt.xlabel("Genre",fontsize=25)
plt.ylabel("Range",fontsize=25)

plt.show()
```

#  This is the second  csv file which we can call this as a movierat which describe the rating of the movies with the imdb title 

```python
movierat=pd.read_csv("IMDb ratings.csv")
movierat
```

# **Cleaning**

```python
movierat.dropna(axis=0,how="any",inplace=True)
movierat.dropna(axis=1,how="any",inplace=True)
```

# Merging Two files movie1 and movierat ? 

```python
da=pd.merge(movie1,movierat,on="imdb_title_id")
india=da[da["country"]=="India"]
india
```

# Cleaning the dataFrame which is made up of two csv files

```python
da.dropna(axis=0,how="any",inplace=True)
```

```python
de=da[da["language"]=="None"]
da.drop(de.index,axis=0,inplace=True)
```

```python
#checking the record in which language is none 
da[da["language"]=="None"]
```

# Q4 Top 10 language used in the movie making in india

```python
#splitting the columns and get the result in the dic format in this function

d={}
def fun(lan):
    if type(lan)!=float:
        
        s=lan.split(",")
        
        for i in range(len(s)):
            s[i]=s[i].replace(" ","")
            if s[i] not in d:
                d[s[i]]=1
            else:
                d[s[i]]=d[s[i]]+1
    return d

india["language"].apply(fun)
```

# plotting the graph

```python
#getting the keys and values and convert into the data frame 
l3=d.keys()
l4=d.values()


#dataframe

indiachart=pd.DataFrame(l4,l3,columns=["values"])
indiachart1=indiachart.sort_values("values",ascending=False)[:10]
indexing=list(indiachart1.index)

plt.figure(figsize=(12,10))
explode=[.1]*10

plt.pie(indiachart1["values"],labels=indexing,autopct='%1.1f%%',explode=explode)
plt.title("Top 10 language in which most of the movies are made ",fontsize=25,color="red")
plt.legend(title="Language")

plt.show()



```

# Q5)Top 10 country in which Maximum No of Movies  are made

```python
#splitting the genre columns and convert into the dictionary format 
d={}
def fun(country):
     
    s=country.split(",")
    for i in range(len(s)):
        s[i]=s[i].replace(" ","")
        if s[i] not in d:
            d[s[i]]=1
        else:
            d[s[i]]=d[s[i]]+1
    return d
    
da["country"].apply(fun)
```

```python
#getting the keys in l3 and values in l4 then make dataframe 
l3=d.keys()
l4=d.values()

#dataframe
world=pd.DataFrame(l4,l3,columns=["Values"])
world1=world.sort_values("Values",ascending=False)[:10]
indexing=list(world1.index)

plt.figure(figsize=(12,10))
explode=[.1]*10


plt.pie(world1,explode=explode,labels=indexing,autopct='%1.1f%%')
plt.legend(title="Country")

plt.title("Top 10 country in which Maximum No of Movies  are made",color="red",fontsize=25)
plt.show()
```
